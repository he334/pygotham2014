```python
from mpl_toolkits.basemap import Basemap, cm
from matplotlib import rcParams
from matplotlib.collections import LineCollection
from matplotlib import cm
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
from shapely.geometry import Polygon as ShapelyPoly
import sys

lllat = 40.732531; urlat = 40.93709; lllon = -74.02113; urlon=-73.820241

m = Basemap(projection='merc', llcrnrlat=lllat, urcrnrlat=urlat, llcrnrlon=lllon,
   urcrnrlon=urlon, resolution='l')

fig = plt.figure(figsize=(12, 12))

def plotneighborhoodlabels(ax):
  nhoods = 'data/nyc_nhoodsnames20101/NHoodNameCentroids'
  nh = ['Central Harlem','Morningside Heights','East Harlem','Washington Heights',
      'Inwood','Marble Hill','Norwood','Bedford Park','KingsBridge']
  m.readshapefile(nhoods, 'nh',drawbounds=False)
  for i, info in enumerate(zip(m.nh_info)):
    x = m.nh[i][0]
    y = m.nh[i][1]
    name = m.nh_info[i]['Name']
    if name in nh:
      plt.text(x,y,name)

def plotdistrict(ax, shapefilename, state, districtnumber):
  m.drawcountries()
  m.readshapefile(shapefilename, 'districts', drawbounds=False)

  for shape, district in zip(m.districts, m.districts_info):
    statename = district['STATENAME']
    district = district['DISTRICT']
    if statename == state:
      p2 = ShapelyPoly(shape)
      if district == districtnumber:
        poly = Polygon(shape, facecolor="green")
        ax.add_patch(poly)
        ax.text(p2.representative_point().x, p2.representative_point().y, district)
      else:
        poly = Polygon(shape, facecolor="gray")
        ax.add_patch(poly)

ax1 = fig.add_subplot(121, title="111th Congress")
plotneighborhoodlabels(ax1)
plotdistrict(ax1, "data/districts111thcongress/districts111", "New York", "15")

ax2 = fig.add_subplot(122, title="113th Congress")
plotneighborhoodlabels(ax2)
plotdistrict(ax2, "data/districts113thcongress/districts113", "New York", "13")
plt.savefig('oldvsnewdistrict.png',dpi=300)
plt.show()
```
